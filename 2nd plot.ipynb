```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_events = pd.read_csv('events.csv')
df_gameinfo = pd.read_csv('ginf.csv')
```

```python
df_events.head()
```

```python
df_events_per_game = df_events.groupby('event_type').size() / df_events.groupby('event_type')['id_odsp'].nunique()
events_per_game = df_events_per_game.tolist()
events = ['Attempt', 'Coner', 'Foul', 'Yellow card', 'Second yellow card', 'Red card', 'Substitution', 'Free kick won', 'Offiside', 'Hand ball', 'Penalty conceded']

plt.figure(figsize=(20,5))
plt.title('Average Events Occurances per game')
plt.bar(events, events_per_game)
plt.show()
```
